**Passo 3: Verificação da normalidade dos dados**

````r
shapiro.test(dados$Altura)
````

O teste de Shapiro-Wilk é usado para verificar se uma amostra de dados segue uma distribuição normal. Os resultados apresentados incluem a estatística de teste (W) e o valor-p associado.

* Estatística de Teste (W): No seu caso, o valor de W é 0.95384.
* Valor-p: O valor-p é 0.214.

O teste de Shapiro-Wilk avalia duas hipóteses:

* Hipótese Nula (H0): Os dados seguem uma distribuição normal (p>0.05).
* Hipótese Alternativa (H1): Os dados não seguem uma distribuição normal (p<0.05).

Portanto, concluímos que os dados da altura podem ser consistentes com uma distribuição normal.


**Passo 4: Realização do teste t para uma amostra**

````r
t.test(dados$Altura, mu=167)
````



* Estatística t (t): O valor é 0.70196.
* Graus de liberdade (df): 29.
* Valor-p (p-value): 0.4883.
* Hipótese Alternativa: A hipótese alternativa sugere que a média real não é igual a 167.
* Intervalo de Confiança de 95% para a Média: Varia de 164.2572 a 172.6095.
* Estimativas da Amostra: A média da amostra é 168.4333.



>O teste-t para uma amostra mostrou que a média de altura da amostra (168,43cm) não é diferente da média de altura nacional (167cm) (t(29) = 0,702; p=0,488).

Observações:

O teste bicaudal é o default; caso deseje unicaudal, necessário incluir: `alternative = "greater"` ou `alternative = "less"`.
Exemplo: `t.test(dados$Altura, mu = 167, alternative = "greater")`

Nesse caso, o teste verificará se é a média amostral é maior que a média testada.

````r
byf.shapiro(Nota_Biol ~ Posicao_Sala, dados)
byf.shapiro(Nota_Fis  ~ Posicao_Sala, dados)
byf.shapiro(Nota_Hist ~ Posicao_Sala, dados)
````

**Lembrando**

```
H0: distribuição dos dados  = normal (p > 0.05)
H1: distribuição dos dados != normal (p <= 0.05)
```

Em todos os casos os valor de p foi maior que 0.05, portanto apresentam distribuição normal.

**Verificação da homogeneidade de variâncias**

Teste de Levene (pacote car)

Observação: Por default, o teste realizado pelo pacote car tem como base a mediana (median). O teste baseado na mediana é mais robusto, porém mudamos para ser baseado na média (comparável ao SPSS).

````r
leveneTest(Nota_Biol ~ Posicao_Sala, dados, center=mean)
````

**Lembrando**

```
H0: as variâncias dos grupos são homogêneas     (p > 0.05)
H1: as variâncias dos grupos não são homogêneas (p <= 0.05)
```

No caso das notas de Biologia, as variâncias entre os grupos são homogêneas.


````r
leveneTest(Nota_Fis  ~ Posicao_Sala, dados, center=mean)
leveneTest(Nota_Hist ~ Posicao_Sala, dados, center=mean)
````

Nos casos das notas de Física e História, as variâncias entre os grupos não são homogêneas.

**Realização do teste t para amostras independentes**

**Lembrando**

```
H0: média do grupo A  = média do grupo B (p > 0.05)
H1: média do grupo A != média do grupo B (p <= 0.05)
```



````r
t.test(Nota_Biol ~ Posicao_Sala, dados, var.equal=TRUE)
````

O número `7.136e-05` significa `0.00007136` em notação decimal. É uma representação conveniente para números pequenos, indicando que o valor é multiplicado por 10 elevado à potência negativa de 5.

Exemplo: `t.test(Nota_Biol ~ Posicao_Sala, dados, var.equal=TRUE, alternative="greater")`

Nesse caso, o teste verificará se é a média do primeiro grupo é maior que a média do segundo. O R está considerando "Frente" como primeiro grupo.

O banco de dados "07-dataset.csv" contém informações de 31 indivíduos tratados com placebo, um anti-hipertensivo já em uso no mercado ou um anti-hipertensivo novo. Verifique se há efeito do tratamento sobre a pressão sistólica e os batimentos cardíacos desses indivíduos. Descreva os resultados de forma apropriada.

````r
if(!require(dplyr)) install.packages("dplyr")
library(dplyr)
if(!require(RVAideMemoire)) install.packages("RVAideMemoire")
library(RVAideMemoire)
if(!require(car)) install.packages("car")
library(car)
if(!require(psych)) install.packages("psych")
library(psych)
if(!require(rstatix)) install.packages("rstatix")
library(rstatix)
if(!require(DescTools)) install.packages("DescTools")
library(DescTools)
````

````r
dados <- read.csv2('https://raw.githubusercontent.com/claudiaanjos/estatistica-aplicada/main/modulos/07-dataset.csv', stringsAsFactors = T)
````

````r
glimpse(dados)
````

**Verificação da normalidade dos dados**

**Shapiro por grupo (pacote RVAideMemoire)**

````r
byf.shapiro(BC ~ Grupo, dados)
````

O grupo Placebo não segue uma distribuição normal.

````r
byf.shapiro(Pressao ~ Grupo, dados)
````

Todos os grupos seguem uma distribuição normal.

**Outra opção: pelo pacote rstatix**

````r
dados %>% group_by(Grupo) %>% shapiro_test(BC)
````

````r
dados %>% group_by(Grupo) %>% shapiro_test(Pressao)
````

**Verificação da homogeneidade de variâncias**

**Teste de Levene (pacote car)**

Observação:
Por default, o teste realizado pelo pacote car tem como base a mediana (median). O teste baseado na mediana é mais robusto. Mudamos para ser baseado na média (comparável ao SPSS).

````r
leveneTest(BC ~ Grupo, dados, center = mean)
````

````r
leveneTest(Pressao ~ Grupo, dados, center = mean)
````

As variâncias dos grupos são homogêneas.

**Verificação da presença de outliers (por grupo)**

**Pacotes dplyr e rstatix**

````r
# Para BC:
dados %>%
  group_by(Grupo) %>%
  identify_outliers(BC)
````

````r
# Pelo boxplot:
boxplot(BC ~ Grupo, data = dados,
        ylab = "Frequência cardíaca (bps)", xlab = "Tratamento")
````

````r
# Para Pressão:
dados %>%
  group_by(Grupo) %>%
  identify_outliers(Pressao)
````

````r
# Pelo boxplot:
boxplot(Pressao ~ Grupo, data = dados,
        ylab = "Pressão arterial (mmHg)", xlab = "Tratamento")
````

Nesse caso a função identificou 4 e o boxplot 3 outliers.

A função do rstatix considera (1.5 * amplitude interquartil) e o boxplot (1.58 * amplitude interquartil), por isso a diferença.

**Realização da ANOVA**

Não estamos na condição ideal de ANOVA de uma via, pois temos outliers e não temos distribuição normal em um dos testes.

Mas seguiremos para fins educativos.

**Lembrando**


```
H0: média do grupo A = média do grupo B = média do grupo C (p > 0.05)
H1: há pelo menos uma diferença entre as médias dos grupos (p <= 0.05)
```



````r
# Criação do modelo para BC
anova_BC <- aov(BC ~ Grupo, dados)
summary(anova_BC)
````

````r
# Criação do modelo para Pressão
anova_Pressao <- aov(Pressao ~ Grupo, dados)
summary(anova_Pressao)
````

Temos pelo menos uma diferença entre as médias dos grupos tanto para BC quanto para Pressão.

**Análise post-hoc**

**Pacote DescTools**

Temos pelo menos uma diferença entre as médias dos grupos. Mas qual grupo é diferente? Utilizando os testes post-hocs.

Post-hocs permitidos: "hsd", "bonferroni", "lsd", "scheffe", "newmankeuls", "duncan".

````r
# Uso do Duncan
PostHocTest(anova_BC, method = "duncan")
````

* AH Padrao - AH Novo: p = 0.1942 (os grupos não são diferentes)    
* Placebo   - AH Novo: p = 0.0027 (existe diferença)
* Placebo   - AH Padrao: p = 6.6e-05 (existe diferença)

````r
# Uso do TukeyHSD
PostHocTest(anova_BC, method = "hsd")
````

* AH Padrao - AH Novo: os grupos não são diferentes    
* Placebo   - AH Novo: existe diferença
* Placebo   - AH Padrao: existe diferença

````r
# Uso do Bonferroni
PostHocTest(anova_BC, method = "bonf")
````

* AH Padrao - AH Novo: os grupos não são diferentes    
* Placebo   - AH Novo: existe diferença
* Placebo   - AH Padrao: existe diferença

````r
# É o mesmo que:
pairwise.t.test(dados$BC, dados$Grupo, p.adj="bonferroni", paired=F)
````

**Exemplo de como resumir em uma tabela mais de um post-hoc**

````r
round(
  cbind(duncan = PostHocTest(anova_BC, method="duncan")$Grupo[,"pval"],
        bonf   = PostHocTest(anova_BC, method="bonf")$Grupo[,"pval"],
        hsd    = PostHocTest(anova_BC, method="hsd")$Grupo[,"pval"]),
  6)
````

````r
# Uso do TukeyHSD para a Pressão:
PostHocTest(anova_Pressao, method = "hsd")
````

* AH Padrao - AH Novo: os grupos não são diferentes    
* Placebo   - AH Novo: os grupos não são diferentes
* Placebo   - AH Padrao: existe diferença

**Análise descritiva dos dados**

````r
describeBy(dados$BC, group = dados$Grupo)
````

````r
describeBy(dados$Pressao, group = dados$Grupo)
````

**Outra opção: pelo pacote rstatix**

````r
dados %>% group_by(Grupo) %>% get_summary_stats(BC, Pressao)
````

> A ANOVA de uma via mostrou que há efeito do tratamento sobre a média de pressão arterial [F(2,28) = 5,521; p = 0,0095]. O post-hoc Tukey HSD mostrou que há diferenças entre o grupo placebo e o grupo anti-hipertensivo padrão, mas não entre os demais grupos.

<br>

O banco de dados 16-dataset.csv contém informações de 30 indivíduos tem efeito sobre os seus salários, controlando esse efeito pela idade. Descreva os resultados.

````r
if(!require(dplyr)) install.packages("dplyr")
library(dplyr)
if(!require(rstatix)) install.packages("rstatix")
library(rstatix)
if(!require(car)) install.packages("car")
library(car)
if(!require(ggplot2)) install.packages("ggplot2")
library(ggplot2)
if(!require(multcomp)) install.packages("multcomp")
library(multcomp)
if(!require(emmeans)) install.packages("emmeans")
library(emmeans)
````

````r
dados <- read.csv2('https://raw.githubusercontent.com/claudiaanjos/estatistica-aplicada/main/modulos/16-dataset.csv', stringsAsFactors = T)
````

````r
head(dados)
````

````r
glimpse(dados)
````

Quero saber se o grau de instrução tem efeito no salário, apesar da idade:

* VD: Salário
* VI: Grau de instrução
* Covariável: Idade

<br>


**Verificar se há efeito da VI sobre a covariável (cov ~ VI)**

* Pressuposto: "independência entre a VI e a covariável"
* Se rompido: não há outro modelo - é um problema de delineamento

````r
mod_cov <- aov(Idade ~ Grau_Instrucao, data = dados)
summary(mod_cov)
````


```
H0: não há efeito do grau de instrução sobre a idade (p >  0,05)
H1: há efeito do grau de instrução sobre a idade     (p <= 0,05)
```

O pressuposto foi atendido.

<br>

**Verificar se a relação entre a covariável e a VD é linear (VD ~ cov)**

````r
ggplot(data = dados, aes(x = Idade, y = Salario, group = Grau_Instrucao,
                         color = Grau_Instrucao)) +
  geom_point(size = 2) +
  xlab('Idade') +
  ylab('Salário') +
  geom_smooth(method = "lm", se = FALSE, size = 0.5)
````

<br>

**Verificar se o efeito da covariável é o mesmo para todos níveis da VI (VD ~ VI*cov)**

* Pressuposto: "homogeneidade dos parâmetros de regressão"
* Compara as inclinações das retas para cada grupo da VI


````r
mod_int <- aov(Salario ~ Grau_Instrucao*Idade, data = dados)
Anova(mod_int, type = 'III')
````

Utilizamos a soma dos quadrados do tipo 3.

Para saber se existe a homogeneidade dos parâmetros de regressão, a interação entre o grau de instrução e idade não podem ser estatisticamente significativa.

```
H0: há homogeneidade dos parâmetros de regressão     (p >  0,05)
H1: não há homogeneidade dos parâmetros de regressão (p <= 0,05)
```

Olhando a penúltima linha da saída, o pressuposto foi atendido.

<br>

**Verificar se há homogeneidade de variâncias (VD ~ VI)**
* Se rompido: versão robusta da ANCOVA

````r
leveneTest(Salario ~ Grau_Instrucao, center = mean, data = dados)
````

```
H0: as variâncias dos grupos são homogêneas     (p >  0,05)
H1: as variâncias dos grupos não são homogêneas (p <= 0,05)
```

Pressuposto atendido.

<br>


**Passo 7: Ajustar o modelo de ANCOVA (VD ~ cov + VI)**
* Se os resultados forem avaliados pelo tipo I da soma dos quadrados é obrigatório que a covariável seja inserida antes no modelo

````r
options(contrasts = c("contr.sum", "contr.poly"))

mod_ANCOVA <- aov(Salario ~ Idade + Grau_Instrucao, data = dados)
Anova(mod_ANCOVA, type = 'III')
````

Existe efeito grau de intrução no salário, controlado pela idade?

```
H0: não há efeito (p >  0,05)
H1: há efeito     (p <= 0,05)
```

Vamos testar outros pressupostos, que só é possível após a contrução do modelo.

<br>

**Verificar a normalidade dos resíduos**
* Se rompido: versão robusta da ANCOVA

````r
shapiro.test(mod_ANCOVA$residuals)
````

**Verificar se há homocedasticidade e outliers**

````r
boxplot(mod_ANCOVA$residuals)
````

````r
par(mfrow=c(1,2))
plot(mod_ANCOVA, which=c(1,3))
````

Para existir a homocedasticidade, a distribuição de pontos devem estar retangular (por todas as áreas).

````r
leveneTest(mod_ANCOVA$residuals ~ dados$Grau_Instrucao)
````

**Realização das comparações entre grupos**

````r
# Pelo pacote multcomp (múltiplas comparações)
posthoc <- glht(mod_ANCOVA, linfct = mcp(Grau_Instrucao = "Tukey"))
summary(posthoc)
````

Outra opção: Dunnett

````r
# Pelo rstatix
comparacoes <- dados %>% emmeans_test(Salario ~ Grau_Instrucao, covariate = Idade,
                       p.adjust.method = "bonferroni")

comparacoes
````


**Obtenção das médias ajustadas**

Ajusta para o efeito da covariável


````r
# Opção 1:
medias_ajustadas <- emmeans(mod_ANCOVA, ~ Idade:Grau_Instrucao) #(~cv:VI)
medias_ajustadas
````

````r
# Opção 2:
get_emmeans(comparacoes)
````

````r
# Médias reais
dados %>% group_by(Grau_Instrucao) %>%
  get_summary_stats(Salario, type = "mean_sd")
````

> O modelo de ANOVA mostrou que há efeito do grau de instrução sobre o salário [F(2, 26) = 4,54; p = 0,020], quando controlado pela idade do indivíduo. As comparações entre pares mostraram que há diferença entre os salários do grupo "Ensino médio" e do grupo "Ensino Superior", sendo que esse último grupo apresenta, em média, salários maiores.

**Passo 5: Construção do modelo da ANOVA com medidas repetidas (pacote: ez)**

````r
mod.ANOVA <- ezANOVA(data = dadosl,
                     dv       = .(TG),
                     wid      = .(ID),
                     within   = .(Tempo),
                     between  = .(Genero),
                     detailed = TRUE,
                     type     = 3)
````

* dv = variável dependente
* wid = variável de identificação do sujeito
* within = variável independente de medidas repetidas
* between = variável independente entre sujeitos
* between_covariates = covariável
* type = tipo da soma dos quadrados (default é o tipo II, tipo III é o padrão no SPSS)

<br>

**Passo 5: Analisando os resultados do modelo**


````r
mod.ANOVA
````
